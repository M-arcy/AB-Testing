<h2 style="color:#005395;">Scenario</h2>
<p style="color:#0090DA;">
A fast-food chain plans to add a new item to its menu. However, they are still undecided between three possible marketing campaigns for promoting the new product. In order to determine which promotion has the greatest effect on sales, the new item is introduced at locations in several randomly selected markets. A different promotion is used at each location, and the weekly sales of the new item are recorded for the first four weeks.
</p>

<h2 style="color:#005395;">Goal</h2>
<p style="color:#FDB813;">
<strong>Evaluate A/B testing results and decide which marketing strategy works the best.</strong>
</p>

  <li><strong>LocationID:</strong> unique identifier for store location</li>

With this data, I'll will answer the question "Is there a difference in the number of reviews among the three promotions?"

In order to answer this question, the x, or independent variable, will be the promotion categories, which have three levels: one, two and three. The y, or dependent variable, will be sales. As with all ANOVAs, the IV will be categorical, and the DV will be continuous.

```python
ff.info()  #checking data type of sales and promotions - are they integers?
```

Bartlett's test is for when your data is normally distributed, and Fligner's test is for when your data is non-parametric. No matter which test you are using, you are looking for a non-significant test. The null hypothesis for both of these is that the data has equal variance, so you'd like to have a p value of > .05.

```python
sns.histplot(ff['SalesInThousands']) #somewhat of a bimodal distribution, and perhaps a bit positively skewed, but I'll proceed as if normally distributed for purposes of this notebook. 
```

```python
scipy.stats.bartlett(ff['SalesInThousands'], ff['Promotion']) #violated homogeneity
```

```python
scipy.stats.fligner(ff['SalesInThousands'], ff['Promotion'])
```

Since p value > .05 here, homogeneity of variance is not violated. In the interest of time, I'll continue in Python with this, but R is easier to correct for this. Sample size > 20 is met, so all assumptions are met for purposes of this notebook. 

```python
stats.f_oneway(ff['SalesInThousands'][ff['Promotion']==1],
                    ff['SalesInThousands'][ff['Promotion']==2],
                    ff['SalesInThousands'][ff['Promotion']==3])
```

interpretation: Given the p-value is much smaller than 0.05 (which is the commonly used threshold for statistical significance), we can reject the null hypothesis and conclude that there is a statistically significant difference in sales between at least two of the promotion types.

However, the ANOVA test doesn't tell me which specific promotions are different from each other, just that at least one pair is different. In order to figure out exactly which promotions are significantly different in terms of sales, I'll conduct post-hoc pairwise comparison tests (such as Tukey's HSD test).

```python
postHoc = MultiComparison(ff['SalesInThousands'], ff['Promotion'])
postHocResults = postHoc.tukeyhsd()
print(postHocResults)
```

<div style="color: #005395;">

<h2 style="font-size: 1.5em;">Understanding the Output</h2>

<p style="font-size: 1.2em;">
This output provides us with the mean difference in the number of sales per promotion, plus the confidence interval (lower and upper columns), and whether or not we can reject the null hypothesis that there is no difference at all between groups. If the value in the 'reject' column is True, then this means there was a significant difference in the means between those groups.
</p>

<h2 style="font-size: 1.5em;">So there's a difference; let's see where</h2>

<p style="font-size: 1.2em;">
So, there is a significant difference between some of the sales recorded for each of the three promotions. Let's take a look at the means of each to see where the difference lies.
</p>

</div>


```python
ff.groupby('Promotion').mean()
```

<div style="color: #005395;">  

<h2 style="font-size: 1.5em;">Analysis Results</h2>

<p style="font-size: 1.2em;">
Based on the results of the Tukey's HSD test, we can draw the following conclusions:
</p>

<ul style="font-size: 1.2em; margin-left: 20px;">
    <li>There is a statistically significant difference in the average sales between Promotion 1 and Promotion 2 (p-value < 0.05). Specifically, Promotion 1 is associated with higher sales by an average of about 10.77 thousand compared to Promotion 2.</li>
    <li>No significant difference in the average sales is found between Promotion 1 and Promotion 3 (p-value > 0.05).</li>
    <li>There is a statistically significant difference in average sales between Promotion 2 and Promotion 3 (p-value < 0.05). Specifically, Promotion 3 is associated with higher sales by an average of about 8.04 thousand compared to Promotion 2.</li>
</ul>
